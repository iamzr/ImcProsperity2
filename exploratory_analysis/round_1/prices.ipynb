# Analysis of Round 1 Data

```python
#imports

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

## Get all the data in one place

For round 1 we have two sets of data the price data and the trade data

```python
round_1_data_path = Path(r"/home/zr/code/trading_project/exploratory_analysis/Prosperity2Data/round-1-island-data-bottle")
```

## Price Data

```python
prices_day_0_csv = round_1_data_path/ Path("prices_round_1_day_0.csv")
prices_day_minus_1_csv = round_1_data_path / Path("prices_round_1_day_-1.csv")
prices_day_minus_2_csv = round_1_data_path / Path("prices_round_1_day_-2.csv")
```

```python
day_0 = pd.read_csv(prices_day_0_csv, sep=";")
day_m1 = pd.read_csv(prices_day_minus_1_csv, sep=";")
day_m2 = pd.read_csv(prices_day_minus_2_csv, sep=";")


data = pd.concat([day_m2, day_m1, day_0]).reset_index()
data
```

# Ananalysis the price data

Got all the data together

```python
# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size

products = data['product'].unique()
for product in products:
    product_data = data[data['product'] == product]
    plt.plot(product_data['price'], label=f'Product {product}')

# plt.xlabel('Timestamp')
plt.ylabel('Bid Price')
plt.title('Bid Price over Time for Each Product')
plt.legend()
plt.grid(True)  # Optionally add a grid
plt.show()
```

## Lets look at each seperately

### Amethysts

```python
# Plotting
plt.figure(figsize=(10, 6))  # Set the figure size

products = ["AMETHYSTS"]
for product in products:
    product_data = data[data['product'] == product]
    plt.plot(product_data['bid_price_1'], label=f'Product {product}')

plt.xlabel('Timestamp')
plt.ylabel('Bid Price')
plt.title('Bid Price over Time for Amethysts')
plt.legend()
plt.grid(True)  # Optionally add a grid
plt.ylim(9980, 10020)

# Add in a line for mean
plt.axhline(y = 9996.58940 , color = 'r', linestyle = '-') 
  
# rendering the plot 
plt.show()

plt.show()
```

From the above we can clearly see that the price of a amethyssts fluctates around a mean

```python
product = "AMETHYSTS"
product_data = data[data['product'] == product]
product_data.describe()
```

We can see the mean for "bid_price_1" is 9996.58940

We can set some upper bounds and lower bounds (bollinger bands) at 2 stds from the mean

```python
mean = product_data["bid_price_1"].mean()
std = product_data["bid_price_1"].std()

k = 2

upper_bound = mean + (k * std)
lower_bound = mean - (k * std)

print(upper_bound, mean, lower_bound)
```
